```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_excel("C:/Users/Najeeb ULLAH/Downloads/Apple historical_data (1).xlsx")
```

```python
df
```

```python
x=df[['Close','High','Low']]
```

```python
x
```

```python
y=df['Close']
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
LinearRegression=LinearRegression()
```

```python
LinearRegression.fit(xtrain,ytrain)
```

```python
predict=LinearRegression.predict(xtest)
```

```python
mean_squared_error(ytest,predict)
```

```python
model=SGDRegressor(max_iter=1000,alpha=0.001)
```

```python
model.fit(xtrain,ytrain)
```

```python
predict=model.predict(xtest)
```

```python
mean_squared_error(ytest,predict)
```
